API DE DEMANDA ELECTRICA

```python
cod_comunidades = {'Ceuta': 8744,
                        'Melilla': 8745,
                        'Andalucía': 4,
                        'Aragón': 5,
                        'Cantabria': 6,
                        'Castilla - La Mancha': 7,
                        'Castilla y León': 8,
                        'Cataluña': 9,
                        'País Vasco': 10,
                        'Principado de Asturias': 11,
                        'Comunidad de Madrid': 13,
                        'Comunidad Foral de Navarra': 14,
                        'Comunitat Valenciana': 15,
                        'Extremadura': 16,
                        'Galicia': 17,
                        'Illes Balears': 8743,
                        'Canarias': 8742,
                        'Región de Murcia': 21,
                        'La Rioja': 20}


lista_años=[2019,2020,2021]
```

```python
headers={"Accept": "application/json",
"Content-Type": "application/json",
"Host": "apidatos.ree.es"}

ruta = r'C:\Users\jaime\Desktop\Thepower\Día15\laboratorio-modulo5-leccion01-etl\datos\demanda'
os.makedirs(ruta, exist_ok=True)

for anio in lista_años:
    for nombre, ccaa in cod_comunidades.items():
        
        url_demanda=f"https://apidatos.ree.es/es/datos/demanda/evolucion?start_date={anio}-01-01T00:00&end_date={anio}-12-31T23:59&time_trunc=month&geo_trunk=electric_system&geo_limit=ccaa&geo_ids={ccaa}"

        response = requests.get(url_demanda, headers=headers)
        responsa= response.json()   
        
        df_demanda= pd.DataFrame(responsa['included'][0]['attributes']['values']).to_csv(os.path.join(ruta, f'{nombre}_{anio}.csv'))
```

```python
headers={"Accept": "application/json",
"Content-Type": "application/json",
"Host": "apidatos.ree.es"}

ruta = r'C:\Users\jaime\Desktop\Thepower\Día15\laboratorio-modulo5-leccion01-etl\datos\oferta'
os.makedirs(ruta, exist_ok=True)

for anio in lista_años:
    for nombre, ccaa in cod_comunidades.items():
        
        url_oferta=f"https://apidatos.ree.es/es/datos/generacion/estructura-renovables?start_date={anio}-01-01T00:00&end_date={anio}-12-31T23:59&time_trunc=month&geo_trunk=electric_system&geo_limit=ccaa&geo_ids={ccaa}"

        response = requests.get(url_oferta, headers=headers)
        if response.status_code != 200:
            continue
        else:
            responsa= response.json()   
            
            df_demanda= pd.DataFrame(responsa['included'][0]['attributes']['values']).to_csv(os.path.join(ruta, f'{nombre}_{anio}.csv'))
```
